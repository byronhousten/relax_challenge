```python
%matplotlib inline
import datetime
import numpy as np
import pandas as pd
```

```python
users = pd.read_csv("takehome_users.csv", encoding='latin1')
```

```python
engagement = pd.read_csv("takehome_user_engagement.csv", encoding='latin1')
```

```python
users.last_session_creation_time = pd.to_datetime(users.last_session_creation_time, unit='s')
```

```python
users.set_index('object_id', inplace=True)
```

```python
users.head()
```

```python
engagement.head()
```

```python
engagement.time_stamp = pd.to_datetime(engagement.time_stamp)
```

```python
login_count = pd.DataFrame(engagement.groupby(by='user_id').size(), columns=['login_count'])
```

```python
users = users.join(login_count)
```

```python
users.login_count.fillna(0, inplace=True)
```

```python
users['invited_by_user'] = np.where(users.invited_by_user_id.isnull(), 0, 1)
```

```python
users['guest_invite'] = np.where(users.creation_source == 'GUEST_INVITE', 1, 0)
users['org_invite'] = np.where(users.creation_source == 'ORG_INVITE', 1, 0)
users['personal_projects'] = np.where(users.creation_source == 'PERSONAL_PROJECTS', 1, 0)
users['signup_google'] = np.where(users.creation_source == 'SIGNUP_GOOGLE_AUTH', 1, 0)
```

```python
users['adopted_user'] = pd.Series()
```

```python
def check_logins(user_timestamps):
    if len(user_timestamps) == 3:
        if ((user_timestamps.iloc[1] - user_timestamps.iloc[0]) <= datetime.timedelta(days=7)) & \
           ((user_timestamps.iloc[2] - user_timestamps.iloc[0]) <= datetime.timedelta(days=7)):
                return 1
        else:
            return 0
    else:
        if ((user_timestamps.iloc[1] - user_timestamps.iloc[0]) <= datetime.timedelta(days=7)) & \
           ((user_timestamps.iloc[2] - user_timestamps.iloc[0]) <= datetime.timedelta(days=7)):
                return 1
        else:
            return check_logins(user_timestamps.iloc[1:])

for id in users.index:
    print("User id: ", id)
    if users.ix[id, 'login_count'] >= 3:
        user_timestamps = engagement.time_stamp[engagement.user_id == id].dt.date.sort_values()
        users.ix[id, 'adopted_user'] = check_logins(user_timestamps)
    else:
        users.ix[id, 'adopted_user'] = 0  
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
feature_var = ['opted_in_to_mailing_list', 'enabled_for_marketing_drip', 'login_count', \
           'guest_invite', 'org_invite', 'personal_projects', 'signup_google']
x = users[feature_var].values
y = users['adopted_user'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, stratify=y)
```

```python
model = GridSearchCV(LogisticRegression(max_iter=1000, n_jobs=2), 
                     param_grid={'C': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100]}, 
                     scoring='accuracy', cv=StratifiedKFold(n_splits=5))
```

```python
model.fit(x_train, y_train)
```

```python
model.cv_results_
```

```python
model2 = LogisticRegression(C=model.best_params_['C'], max_iter=1000, n_jobs=4)
model2.fit(x_train, y_train)
```

```python
model2.coef_
```

```python
model2.intercept_[0]
```

```python
pred = model2.predict(x_test)
```

```python
accuracy_score(y_true=y_test, y_pred=pred)
```

```python
confusion_matrix(y_true=y_test, y_pred=pred)
```

```python
print(classification_report(y_true=y_test, y_pred=pred, digits=2))
```

```python
coef = [n for n in model2.coef_[0]]
coef.append(model2.intercept_[0])
```
